```python
import os
import re
import glob
os.getcwd()
from google.colab import drive

drive.mount("/content/drive")
```

/content/drive/My Drive/phd_thesis_data/deep_city/paired_datasets/w-s/train/

```python
PATH = "/content/drive/My Drive/phd_thesis_data/deep_city/pix2pix/compiled/combined/thumbnails/w-m/ebird_raw/"#@param{type:'string'}
PATH = PATH.rstrip("/") + "/"

# Supported image formats
patterns = ["*.jpg", "*.jpeg", "*.png", "*.JPG", "*.JPEG", "*.PNG"]
```

```python
# Collect all image files
files = []
for p in patterns:
    files.extend(glob.glob(PATH + p))

print("Found:", len(files), "images")


# Extract numeric part for sorting
def extract_number(fname):
    nums = re.findall(r"\d+", os.path.basename(fname))
    return int(nums[-1]) if nums else float('inf')

# Sort images based on numbers in filenames
files_sorted = sorted(files, key=extract_number)

START_NUM = 0 #@param{type:'integer'}

# Rename files using 4-digit numbers starting from START_NUM
for offset, old_file in enumerate(files_sorted):
    new_num = START_NUM + offset
    ext = old_file.split(".")[-1]
    new_name = f"{new_num:04d}.{ext}"
    new_path = os.path.join(PATH, new_name)
    os.rename(old_file, new_path)
```

```python
print("Renaming complete.")
print(os.listdir(PATH))
```
