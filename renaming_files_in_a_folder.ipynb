```python
import os
import re
import glob
import shutil
from PIL import Image
os.getcwd()
from google.colab import drive

drive.mount("/content/drive")
```

```python
# ========= User Params =========
PATH = "/content/drive/My Drive/plants-facades_dataset/facade/"  #@param {type:"string"}
START_NUM = 295  #@param {type:"number"}

PATH = PATH.rstrip("/") + "/"

# ========= STEP 1: READ ALL FILES FROM DRIVE AND CACHE LOCALLY =========
LOCAL = "/content/local_cache_clean/"
OUT_LOCAL = "/content/local_output_clean/"

shutil.rmtree(LOCAL, ignore_errors=True)
shutil.rmtree(OUT_LOCAL, ignore_errors=True)

os.makedirs(LOCAL, exist_ok=True)
os.makedirs(OUT_LOCAL, exist_ok=True)

files_drive = glob.glob(PATH + "*")

for f in files_drive:
    shutil.copy(f, LOCAL)

print("Original file count:", len(files_drive))
```

```python
# ========= STEP 2: SORT FILES (alphabetically to avoid number conflicts) =========
patterns = ["*.jpg", "*.jpeg", "*.png", "*.JPG", "*.JPEG", "*.PNG"]
files = []
for p in patterns:
    files.extend(glob.glob(LOCAL + p))

files_sorted = sorted(files, key=lambda x: os.path.basename(x).lower())

print("Files sorted for processing:", len(files_sorted))
```

```python
# ========= STEP 3: CONVERT OR COPY TO NEW PNGS =========
for offset, old_file in enumerate(files_sorted):
    new_num = START_NUM + offset
    new_name = f"{new_num:04d}.png"
    new_path = os.path.join(OUT_LOCAL, new_name)

    try:
        img = Image.open(old_file)
        img = img.convert("RGB")
        img.save(new_path, "PNG")
    except:
        shutil.copy(old_file, new_path)

print("Converted output count:", len(os.listdir(OUT_LOCAL)))
```

```python
# ========= STEP 4: CLEAN DRIVE DIRECTORY FULLY =========
for f in glob.glob(PATH + "*"):
    os.remove(f)

print("Drive folder cleaned.")
```

```python
# ========= STEP 5: MOVE NEW FILES BACK =========
for f in glob.glob(OUT_LOCAL + "*.png"):
    shutil.copy(f, PATH)

final_count = len(os.listdir(PATH))
print("Final file count:", final_count)
```
